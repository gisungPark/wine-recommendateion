```python
# 크롤링 모듈 임포트
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

from bs4 import BeautifulSoup
import time
import collections
from collections import OrderedDict
import csv
```

```python
chrome_driver = 'C:/Users/USER/chromedriver.exe'
driver = webdriver.Chrome(chrome_driver)
driver.implicitly_wait(5) # 응답의 시간 지연
# url = 'https://www.starbucks.co.kr/store/getStore.do?r=24F3U9PI7Q'
url = 'https://www.wine21.com/13_search/wine_list.html?Page=1&PageSize=10&shKeyword=&shOrder=POINT^DESC&shFILTERCNT=1&shWINEPRICE1=10000&shWINEPRICE2=30000#List'
# chrome driver로 해당 페이지가 물리적으로 open
driver.get(url)

#search = driver.find_element_by_class_name('column_detail1')
#print(search)

src = driver.page_source
soup = BeautifulSoup(src)
name = soup.select('div.column_detail1 img')
name=name[1:]
for i in name:
    print(i)

    
```

```python
# 와인을 모을 리스트
#wine_list=[]
```

```python
search=driver.find_elements_by_xpath("//h4[@class='name_ko']/a")


for i in range(len(search)):
# for item in search:
    list = []
    print("*******************")
    time.sleep(1)
    search=driver.find_elements_by_xpath("//h4[@class='name_ko']/a")
    #print(search[i])
    driver.execute_script("arguments[0].click();", search[i])
    
    time.sleep(1)
    src = driver.page_source
    
    # BeautifulSoup 객체로 변환
    soup = BeautifulSoup(src)
    
    #와인식별번호 - outoincreament
    
    #와인사진
    tmp = soup.select('div.column_detail1 > div.thumb > img')
    tmp=str(tmp)
    photo=tmp.split('"')[3]
    print('사진:'+ photo)
    list.append(photo)
    
    #한국와인명
    koname= tmp.split('"')[1]
    print('한글이름:' +koname)
    list.append(koname)
    
    #영어와인명
    ename = soup.select('.name_en')
    ename=str(ename).split('>')[1].split('<')[0]
    print('영어이름:' +ename)
    list.append(ename)
    
    #와인정보가져오기
    wine = soup.select('div.wine_info')
    wine = str(wine)
    
    
    #알콜도수
    if(wine.find('알코올 도수')==True):
        alchol =wine.split('알코올 도수')[1].split('</dd>')[0].split('<dd>')[1]
    else:
        alchol = '13~14도'
    print('알콜도수:'+ alchol)
    list.append(alchol)
    
    #와인종류
    type =wine.split('종 류')[1].split('<dd>')[1].split('\xa0')[0]
    print('와인종류:' +type)
    list.append(type)
    
    #품종식별번호
    pum = wine.split('품 종')[1].split('<a')[1].split('</a')[0].split('>')[1]
    print('품종: '+pum)
    list.append(pum)
    
    #음용온도
    temper=wine.split('음용온도')[1].split('<dd>')[1].split('<')[0]
    print('음용온도:'+temper)
    list.append(temper)
    
    # 와인평점 - 0
    
    print('와인 평점: ')
    list.append('0')
    
    # wine 정보
    info=soup.select('strong')
    info=str(info)
    #print(info)
    # 와인 가격 - 0
    price=info.split('<strong>')[1].split('원')[0]
    print('와인 가격: '+price)
    list.append(price)
    
    #와인연도
    year = info.split('(')[1].split(',')[0]
    print('와인 연도: '+year)
    list.append(year)
    
    #와인설명
    description=soup.select('#MakerNote_wrap')
    description=str(description)
    description=description.split('>')[1].split('<')[0]
    print('와인설명 : '+ description)
    list.append(description)
    
    #와인 판매처
    print('와인 판매처: ')
    
    #당도
    sweet =wine.split('SWEET')[1][0]
    print('당도:'+sweet)
    list.append(sweet)
    
    #산도
    acid = wine.split('ACIDITY')[1][0]
    print('산도:'+acid)
    list.append(acid)
    
    #바디
    body=wine.split('BODY')[1][0]
    print('바디:'+body)
    list.append(body)
    
    #탄닌
    tanin=wine.split('TANNIN')[1][0]
    list.append(tanin)
    
    print('탄닌:'+tanin)
    time.sleep(3)
   
    #뒤로가기
    driver.back()
    
    print(list)
    wine_list.append(list)
```
